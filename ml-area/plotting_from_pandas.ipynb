```python
df_car_sales['Price'] = df_car_sales['Price'].replace('[\$\,\.]', '',regex=True).astype(int)//100
```

```python
df_car_sales
```

```python
df_car_sales['Total Sales'] = df_car_sales['Price'].cumsum()
```

```python
df_car_sales
```

```python
df_car_sales.plot(x='Sales Date', y='Total Sales');
```
